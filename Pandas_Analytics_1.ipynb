```python
import pandas as pd
import json
```

```python
with open('/content/t20_wc_match_results.json') as f:
  data=json.load(f)
df_match=pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.rename({'scorecard':'match_id'},inplace=True,axis=1)
df_match
```

```python
df_match.to_csv('/content/t20_wc_match_results.csv')
```

```python
with open('/content/t20_wc_batting_summary.json') as f:
  data=json.load(f)
  all_records=[]
  for i in data:
    all_records.extend(i['battingSummary'])
df_batting=pd.DataFrame(all_records)
df_batting.tail()
```

```python
df_batting['out/not_out']=df_batting.dismissal.apply(lambda x:'out' if len(x)>0 else 'not_out')
df_batting
```

```python
df_batting.drop(columns=['dismissal'],inplace=True)
df_batting
```

```python
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:x.replace('',''))
df_batting['batsmanName']=df_batting['batsmanName'].apply(lambda x:x.replace('\xa0',''))
df_batting
```

```python
match_ids={}
for index,row in df_match.iterrows():
  key1=row['team1']+' Vs '+row['team2']
  key2=row['team2']+' Vs '+row['team1']
  match_ids[key1]=row['match_id']
  match_ids[key2]=row['match_id']
match_ids
```

```python
df_batting['match_id']=df_batting['match'].map(match_ids)
df_batting
```

```python
df_batting.to_csv('/content/t20_wc_batting_summary.csv',index=False)
```

```python
with open('/content/t20_wc_bowling_summary.json') as f:
    data = json.load(f)
    all_records = []
    for rec in data:
        all_records.extend(rec['bowlingSummary'])
all_records[:2]
```

```python
df_bowling = pd.DataFrame(all_records)
print(df_bowling.shape)
df_bowling.head()
```

```python
df_bowling['match_id'] = df_bowling['match'].map(match_ids)
df_bowling.head()
```

```python
df_bowling.to_csv('/content/t20_wc_bowling_summary.csv', index = False)
```
